```python
def calculate_vif(X, threshold=0.0):
    X_with_const = add_constant(X)
    vif_data = pd.DataFrame()
    vif_data["feature"] = X_with_const.columns
    vif_data["VIF"] = [
        variance_inflation_factor(X_with_const.values, i)
        for i in range(X_with_const.shape[1])
    ]
    # Bỏ cột 'const' khỏi kết quả
    vif_data = vif_data[vif_data['feature'] != 'const']

    # Lọc nếu vượt threshold
    return vif_data[vif_data['VIF'] > threshold].sort_values(by='VIF', ascending=False)

calculate_vif(X)
```

> các biến chỉ có biến region là khả năng sẽ xảy ra đa cộng tuyến, tuy nhiên kiểm tra thì kết quả <2, vì vậy có thể sử dụng được. Nếu không thấy an toàn khi sd biến có thể cân nhắc loại bỏ (biến tương quan thấp với target)

#### 4.1 Mô hình hồi quy tuyến tính

```python
def run_linear_cv(data, target_col='charges', n_splits=5, return_score=False):
    # 1. Chuẩn bị dữ liệu
    X = data.drop(columns=[target_col]).copy()
    y = data[target_col].values

    # 2. KFold CV
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=2025)
    mse_train, mse_test = [], []
    mae_train, mae_test = [], []
    r2_train, r2_test = [], []

    # 3. Vòng lặp CV
    for train_idx, test_idx in kf.split(X):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        model = LinearRegression()
        model.fit(X_train, y_train)

        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

        # MSE
        mse_train.append(mean_squared_error(y_train, y_train_pred))
        mse_test.append(mean_squared_error(y_test, y_test_pred))

        # MAE
        mae_train.append(mean_absolute_error(y_train, y_train_pred))
        mae_test.append(mean_absolute_error(y_test, y_test_pred))

        # R²
        r2_train.append(r2_score(y_train, y_train_pred))
        r2_test.append(r2_score(y_test, y_test_pred))

    # 4. Trung bình
    avg_train_mse = np.mean(mse_train)
    avg_test_mse = np.mean(mse_test)
    avg_train_rmse = np.sqrt(avg_train_mse)
    avg_test_rmse = np.sqrt(avg_test_mse)
    avg_train_mae = np.mean(mae_train)
    avg_test_mae = np.mean(mae_test)
    avg_train_r2 = np.mean(r2_train)
    avg_test_r2 = np.mean(r2_test)

    # 5. Nếu dùng để tối ưu hóa
    if return_score:
        return avg_test_r2

    # 6. In kết quả
    print("\n Linear Regression Results (K-Fold CV)")
    print(f"Average Train MSE: {avg_train_mse:.4f}, Test MSE: {avg_test_mse:.4f}")
    print(f"Average Train RMSE: {avg_train_rmse:.4f}, Test RMSE: {avg_test_rmse:.4f}")
    print(f"Average Train MAE : {avg_train_mae:.4f}, Test MAE : {avg_test_mae:.4f}")
    print(f"Average Train R²  : {avg_train_r2:.4f}, Test R²  : {avg_test_r2:.4f}")
```

```python
run_linear_cv(df)
```

#### **Kết luận:**

* R² không quá cao, tuy nhiên việc ưu tiên hơn hết là sự ổn định (Đã thêm K-fold).
* Về mặt kỹ thuật thì đây là một kết quả chấp nhận được, mô hình này thể hiện tốt với insurance dataset, hướng đi tiếp theo có 2 hướng, sử dụng mô hình phi tuyến của ML cơ bản, hoặc sử dụng các mô hình black box của DL.
* Một hướng đi khác có thể cân nhắc là sử dụng mô hình hồi quy tuyến tính có trọng số (R1, R2) tuy nhiên cần cân nhắc sau khi đã tính được đa cộng tuyến toàn mô hình, mô hình có vẻ ổn, có thể sử dụng với mục đích tham khảo (Tôi nghĩ nó sẽ khá tương đồng với Linear).
* Về mặt thống kê: tìm ra được tính tương quan của toàn bộ X -> y, biết được biến nào sẽ ảnh hưởng nhất tới biến "target" (hút thuốc), theo sau đó là các biến như tuổi tác, chỉ số BMI.
* Trực quan được các biến trong dữ liệu, làm sạch cũng như vẽ được phân phối các biến.
